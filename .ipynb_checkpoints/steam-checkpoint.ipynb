## How many users both buy  Counter-Strike Global Offensive - Dota 2

```python
df3 = df[((df["game_name"] == "Dota 2") | (df["game_name"] == "Counter-Strike Global Offensive")) & (df["behavior_name"] == "purchase")].groupby("user_id").sum().sort_values("behavior", ascending = False)
df3[(df3["behavior"] == 2.0)]
```

## How many users both buy  Team Fortress 2 - Dota 2

```python
df4 = df[((df["game_name"] == "Dota 2") | (df["game_name"] == "Team Fortress 2")) & (df["behavior_name"] == "purchase")].groupby("user_id").sum().sort_values("behavior", ascending = False)
df4[(df4["behavior"] == 2.0)]
```

## How many users both buy  Counter-Strike - Counter-Strike Global Offensive

```python
df5 = df[((df["game_name"] == "Counter-Strike") | (df["game_name"] == "Counter-Strike Global Offensive")) & (df["behavior_name"] == "purchase")].groupby("user_id").sum().sort_values("behavior", ascending = False)
df5[(df5["behavior"] == 2.0)]
```
